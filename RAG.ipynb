```python
%pip install pandas
import pandas as pd

# -----------------------------
# LOAD DATASET
# -----------------------------
df = pd.read_csv("final_dataset_expanded.csv")

# Handle missing values
df = df.fillna("Not specified")

documents = []

# -----------------------------
# CONVERT ROWS → TEXT DOCUMENTS
# -----------------------------
for _, row in df.iterrows():
    text = f"""
    {row['university name']} in {row['country']} offers the program {row['program']}.

    Admission Requirements:
    - GRE Score: {row['gre score']}
    - TOEFL Score: {row['toefl score']}
    - Minimum CGPA: {row['cgpa']}

    Application Details:
    - Deadline: {row['deadline']}
    - Required Documents: {row['required documents']}

    Financial Information:
    - Tuition Fee: {row['tuition fee']}

    Admission Insight:
    - Chance of Admission: {row['chance of admit']}
    """

    documents.append(text.strip())

print("Total documents created:", len(documents))
print("\n Sample Document:\n")
print(documents[0])


# -----------------------------
# CLEAN TEXT (REMOVE LINE BREAKS)
# -----------------------------
clean_documents = [doc.replace("\n", " ") for doc in documents]


# -----------------------------
# SAVE OUTPUT AS CSV
# -----------------------------
output_df = pd.DataFrame({
    "id": range(len(clean_documents)),
    "content": clean_documents
})

output_df.to_csv("processed_documents.csv", index=False)

print("Processed CSV saved successfully!")


# -----------------------------
# OPTIONAL: SAVE AS TXT
# -----------------------------
with open("processed_documents.txt", "w", encoding="utf-8") as f:
    for doc in documents:
        f.write(doc + "\n\n")

print("TXT file also saved!")
```

```python
%pip install langchain-community langchain-huggingface
%pip install sentence-transformers
%pip install faiss-cpu


from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

# -----------------------------
# STEP 1 (reuse your code)
# -----------------------------
df = pd.read_csv("final_dataset_expanded.csv")
df = df.fillna("Not specified")

documents = []

for _, row in df.iterrows():
    text = f"""
    {row['university name']} in {row['country']} offers the program {row['program']}.

    Admission Requirements:
    - GRE Score: {row['gre score']}
    - TOEFL Score: {row['toefl score']}
    - Minimum CGPA: {row['cgpa']}

    Application Details:
    - Deadline: {row['deadline']}
    - Required Documents: {row['required documents']}

    Financial Information:
    - Tuition Fee: {row['tuition fee']}

    Admission Insight:
    - Chance of Admission: {row['chance of admit']}
    """
    documents.append(text.strip())

# -----------------------------
# STEP 2: Embeddings
# -----------------------------
print("🔄 Creating embeddings...")

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Convert documents into FAISS vector store
vector_db = FAISS.from_texts(documents, embedding_model)

# Save locally (IMPORTANT)
vector_db.save_local("faiss_index")

print("FAISS vector DB created and saved!")
```

```python
import os
from dotenv import load_dotenv

from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from groq import Groq

# -----------------------------
# LOAD ENV VARIABLES
# -----------------------------
load_dotenv()
api_key = os.getenv("GROQ_API_KEY")

if not api_key:
    raise ValueError("GROQ_API_KEY not found in .env file")

# -----------------------------
# LOAD FAISS INDEX
# -----------------------------
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_db = FAISS.load_local(
    "faiss_index",
    embedding_model,
    allow_dangerous_deserialization=True
)

print("FAISS loaded successfully!")


# -----------------------------
# GROQ SETUP
# -----------------------------
client = Groq(api_key=api_key)


# -----------------------------
# RAG FUNCTION
# -----------------------------
def ask_chatbot(query):

    # Step 1: Retrieve relevant documents
    docs = vector_db.similarity_search(query, k=15)

    context = "\n\n".join([doc.page_content for doc in docs])

    # Step 2: Create prompt
    prompt = f"""
    You are an intelligent college admission assistant.

    Use the provided context to answer the question.
    If exact information is not available, analyze similar universities and provide a reasonable estimate.

    Do NOT say "Not available" immediately.
    Try to infer based on similar data.
    If the prompt is out of topic or domain, just don't make assumptions on your own.Just say out of domain.

    Context:
    {context}

    Student Profile:
    {query}

    Answer clearly with reasoning:
    """
    # Step 3: Call Groq LLM
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )

    return response.choices[0].message.content
```
